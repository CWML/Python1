```python
#here we save our function reading in the data file to the variable data

data = np.loadtxt(fname="inflammation-01.csv", delimiter=",")
```

```python
print(data)
```

```python
#what you're seeing here is an n-dimensional array, a special kind of numpy data type
print(type(data))
```

```python
#what kind of data is in our dataset - floats, which are numbers with decimals
print(data.dtype)
```

```python
#syntax for indexing in numpy: data[rows, columns]
print("First value in data:", data[0,0])
```

```python
#syntax for indexing with slices in numpy: 
#data[row_slice_start:row_slice_end, column_slice_start_column_slice_end]

print(data[0:4, 0:10])
```

```python
#use numpy's mean function to get the overall average of the data

print(np.mean(data))
```

```python
#this is called multi-value assignment, making 3 variables separated by commas
#assigned to three functions, separated by commas
#just a way to save typing, we could have easily done each of these separately

maxval, minval, stdval = np.max(data), np.min(data), np.std(data)

print("max inflammation", maxval)
print("min inflammation", minval)
print("standard deviation", stdval)
```

```python
#patient_0 is assigned to particular index of the data
#where 0 represents the first (0) row, and : is a shortcut for all columns
#essentially patient_0 is equivalent to the first row of data

patient_0 = data[0, :]

print("Maximum inflammation for Patient 0:", np.max(patient_0))
```

```python
#here we use the 0 axis which goes down each column 
#we could also use the 1 axis which would go across each row (that would be 40 individual patient averages)
#below calculate the average for each day of the trial

print(np.mean(data, axis=0))
```

```python
#now import our visualization library matplotlib
#matplotlib docs - https://matplotlib.org/

import matplotlib.pyplot as plt
```

```python
#show a heatmap of the data 

image = plt.imshow(data)
plt.show()
```

```python
#create a variable representing the 0 axis (the calculation of all numbers in a column)
#in this case, that means the average of all patients' inflammation numbers per day in the trial
#then plot it as a line chart

ave_inflammation = np.mean(data, axis=0)

plt.plot(ave_inflammation)
plt.show()
```

```python
#instead of just displaying the chart, save it as a png file

ave_inflammation = np.mean(data, axis=0)

plt.plot(ave_inflammation)
plt.savefig("ave_inflammation_chart.png")
```

# Post-Class Exercise

Now that we have one average inflammation per day chart, we want to compare it to the additional data files we have, which are additional rounds of the trial with new cohorts of patients. To do this, we'll need to import a new library and reuse some of our looping and conditional logic skills from before.

```python
#import the glob library, which is a library allowing us to access/manipulate files

import glob
```

The line of code below does several things:
1. It uses `glob.glob()` to access the files in this folder according to a pattern. That pattern is the filename has to start with `inflammation` which is denoted by that word, but can be following by anything else, which is denoted by the the wildcard symbol (`*`), and must end in `.csv`. That means that any filename that starts with `inflammation`, no matter what it's followed by (`-01` or `-12`), and that ends in `.csv` will be found and then converted into a Python list of the filenames.
2. It also uses `sorted()`, a Python function that allows us to sort data types such as lists in order, and by default sorts ascending.
3. It finally uses `print()` to show us the results of the prior two operations. 

```python
print(sorted(glob.glob("inflammation*.csv")))
```

Now that we've gained access to the files, and now that `glob.glob()` has conveniently turned them into a list, we can combine everything we've learned into code that visualizes each file. We'll do that below, step-by-step, as follows:

- Store the sorted filenames in a variable called `filenames`.
- Create a `for` loop where we'll loop through each filename, or item in the `filenames` list.
    - In the loop, first print the filename, so we know which file the chart refers to.
    - Read in the datafile using `np.loadtext()` where the `fname` parameter is equivalent to our loop variable `filename`, i.e., each filename in the list.
    - Create an `ave_inflammation` variable as we did above.
    - Plot `ave_inflammation`.
    - Add stylistic elements with matplotlib, including title, a label for the x-axis, and a label for the y-axis.
    - Show each plot as outputs.

```python
filenames = sorted(glob.glob("inflammation*.csv"))

for filename in filenames:
    print(filename)
    data = np.loadtxt(fname=filename, delimiter=",")
    
    ave_inflammation = np.mean(data, axis=0)
    plt.plot(ave_inflammation)
    
    plt.title("Average inflammation flare-ups per day")
    plt.xlabel("Day")
    plt.ylabel("Average")
    
    plt.show()
```
